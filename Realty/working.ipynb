```python
# Create a simple diagram (PNG + SVG) illustrating "Next Layer of Complexity" in AI-led wealth management
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

# Canvas
fig, ax = plt.subplots(figsize=(12, 8))
ax.axis('off')

# Helper to draw box
def box(x, y, w, h, text, fontsize=10):
    
    rect = FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.02,rounding_size=0.05")
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, text, ha='center', va='center', fontsize=fontsize, wrap=True)

# Central box
box(0.15, 0.75, 0.50, 0.13, "AI lowers run-cost → more entrants\nWhere does the next complexity moat emerge?", fontsize=11)

# Four pillars
box(0.05, 0.50, 0.40, 0.12, "Data rights & provenance\n(lineage, usage rights, auditability)")
box(0.57, 0.50, 0.40, 0.12, "Model risk & evaluation\n(tests, guardrails, explainability)")
box(0.05, 0.25, 0.40, 0.12, "Personalization-at-scale\n(mass-custom portfolios, tax/ESG/risk)")
box(0.57, 0.25, 0.40, 0.12, "Microstructure-aware execution\n(RL/agentic TCA, best-ex proofs)")

# Arrows from central to each pillar
def arrow(xy1, xy2):
    arr = FancyArrowPatch(posA=xy1, posB=xy2, arrowstyle='->', mutation_scale=10, linewidth=1.5)
    ax.add_patch(arr)

# arrow((0.50, 0.75), (0.24, 0.56))  # to Data rights
# arrow((0.50, 0.75), (0.76, 0.56))  # to Model risk
# arrow((0.50, 0.75), (0.24, 0.31))  # to Personalization
# arrow((0.50, 0.75), (0.76, 0.31))  # to Microstructure

# Footer
ax.text(0.5, 0.08, "Next-level edge comes from industrializing these four domains.", ha='center', va='center', fontsize=9)

png_path = "next_complexity_ai_wealth.png"
svg_path = "next_complexity_ai_wealth.svg"
plt.savefig(png_path, dpi=300, bbox_inches='tight')
plt.savefig(svg_path, bbox_inches='tight')
png_path, svg_path
```
